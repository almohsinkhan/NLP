```python
# create data 
text = []
label = []


# positive label data
for file in POS_DIR.iterdir():
    # open file 
    with open(file, encoding='utf8') as f:
        text.append(f.read())
        label.append(1)

# negative label data
for file in NEG_DIR.iterdir():
    # open file 
    with open(file, encoding = 'utf8') as f:
        # append in text and label 
        text.append(f.read())
        label.append(0)
```

```python
import re
def clean_text(text):
    text = text.lower()
    text = text = re.sub(r"<.*?>", "", text)
    text =  re.sub(r"[^a-zA-Z\s]", "", text)

    return text

# clean
text = list(map(clean_text, text))

text[0]
```

```python
text[100]
```

```python
# tokenize the text 
def tokenize(text):
    return text.split()
tokenize_text = list(map(tokenize, text))
```

```python
tokenize_text[0][:100]
```

```python
# create vocubulary 
from collections import Counter

word_counter = Counter()

for tokens in tokenize_text:
    word_counter.update(tokens)
```

```python
word2idx = {
    "<PAD>":0,
    "<UNK>":0
}

for word in word_counter:
    word2idx[word] = len(word2idx)
```

```python
#idx to word 
idx2word = {idx: word for word, idx in word2idx.items()}
```

```python
vocab_size = len(word2idx)
vocab_size
```

```python
# replace word with word id \
encoded_texts = []
for tokens in tokenize_text:
    encoded = []

    for word in tokens:
        encoded.append(word2idx.get(word, word2idx["<UNK>"]))
    encoded_texts.append(encoded)
```

```python
import numpy as np

lengths = np.array([len(review) for review in encoded_texts])

print("Max:", lengths.max())
print("Mean:", lengths.mean())
print("Median:", np.median(lengths))
print("90th percentile:", np.percentile(lengths, 90))
print("95th percentile:", np.percentile(lengths, 95))
print("99th percentile:", np.percentile(lengths, 99))
```

```python
MAX_LEN = 300

# padding function
def pad_sequence(sequence, max_len, pad_value = 0):
    if(len(sequence) > max_len):
        return sequence[:max_len]

    return sequence + [pad_value] * (max_len - len(sequence))
```

```python
print(pad_sequence([5,3,4],6))
```

```python
padded_texts = []

for review in encoded_texts:
    padded = pad_sequence(review, MAX_LEN)
    padded_texts.append(padded)
```

```python
print(padded_texts[0])
```

```python
import torch

X = torch.tensor(padded_texts, dtype=torch.long)

y = torch.tensor(label, dtype=torch.float32)
```

```python
from torch.utils.data import TensorDataset

dataset = TensorDataset(X, y)
```

```python
from torch.utils.data import DataLoader

train_loader = DataLoader(
    dataset,
    batch_size=32,
    shuffle=True
)
```

```python
X, y = next(iter(train_loader))
X.shape, y.shape
```

```python
# build the model 
import torch.nn as nn
class LSTMClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.embedding = nn.Embedding(
            num_embeddings = vocab_size,
            embedding_dim = 128
        )
        # lstm layer
        self.lstm = nn.LSTM(
            input_size=128,
            hidden_size=128,
            batch_first=True
        )

        # linear fully connected layer 
        self.fc = nn.Linear(
            128, 1
        )

    def forward(self, x):
        x =  self.embedding(x)

        out, (hidden, cell) = self.lstm(x)

        out = out[:, -1, :]
        out = self.fc(out)

        return out.squeeze(1)
```

```python
# device 
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu') 
# create model
model = LSTMClassifier().to(device)

# loss function
criterion = nn.BCEWithLogitsLoss()

# optimizer
optimizer = torch.optim.Adam(
    model.parameters(),
    lr=0.001
)
```

```python
model.parameters() 
```

```python
output = model(X.to(device))
output
```

```python
# compute loss 
loss  = criterion(
    output,
    y.to(device)
)
loss
```

```python
loss.backward()
```

```python
epochs = 20

for epoch in range(epochs):
    model.train()
    total_loss = 0

    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()

        outputs = model(X_batch.to(device))

        loss =  criterion(outputs, y_batch.to(device))

        loss.backward()

        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch + 1}, Loss : {total_loss/ len(train_loader):.4f}")
```

```python
model
```

```python
model.eval()

X, y = next(iter(train_loader))

X = X.to(device)
y = y.to(device)

with torch.no_grad():
    pred = model(X)
    prob = torch.sigmoid(pred)
    prediction = (prob >= 0.5).long()

print("Prediction:", prediction)
print("Actual:", y)

accuracy = (prediction == y).float().mean()

print(f"Accuracy: {accuracy.item():.4f}")
```
